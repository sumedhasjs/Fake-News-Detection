```python
truenews['True/Fake']='True'
fakenews['True/Fake']='Fake'
```

```python
news = pd.concat([truenews, fakenews])
news["Article"] = news["title"] + news["text"]
news.sample(frac = 1)
```

```python
import nltk
nltk.download('stopwords')
import string
```

```python
def TextProcessing(sentance):
    RemovedPunc = [char for char in sentance if char not in string.punctuation]
    RemovedPunc = ''.join(RemovedPunc)
    
    # Convert string to lowercase and remove stopwords
    CleanedString = [word for word in RemovedPunc.split() if word.lower() not in stopwords.words('english')]
    return CleanedString
```

```python
news['Clean Text'] = news['Article'].apply(TextProcessing)
```

```python
news.sample(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
BagOfWordsTransformer = CountVectorizer(analyzer=TextProcessing).fit(news['Clean Text'])

print(len(BagOfWordsTransformer.vocabulary_))
```

```python
NewsBOW = BagOfWordsTransformer.transform(news['Clean Text'])
```

```python
print('Shape of Sparse Matrix: ', NewsBOW.shape)
print('Amount of Non-Zero occurences: ', NewsBOW.nnz)
```

```python
sparsity = (100.0 * NewsBOW.nnz / (NewsBOW.shape[0] * NewsBOW.shape[1]))
print('sparsity: {}'.format(round(sparsity)))
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tf_idf_transformer = TfidfTransformer().fit(NewsBOW)
news_tf_idf = tf_idf_transformer.transform(NewsBOW)
print(news_tf_idf.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
fakenews_detect_model = MultinomialNB().fit(news_tf_idf, news['True/Fake'])
```

```python
predictions = fakenews_detect_model.predict(news_tf_idf)
print(predictions)
```

```python
from sklearn.metrics import classification_report
print (classification_report(news['True/Fake'], predictions))
```

```python
from sklearn.model_selection import train_test_split

news_train, news_test, text_train, text_test = train_test_split(news['Article'], news['True/Fake'], test_size=0.3)

print(len(news_train), len(news_test), len(news_train) + len(news_test))
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('bow', CountVectorizer(analyzer=TextProcessing)),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
pipeline.fit(news_train,text_train)
```

```python
predictions = pipeline.predict(news_test)
```

```python
print(classification_report(predictions,text_test))
```
